```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.realtor.com/international/my//'
```

```python
# Headers for requests
Headers = ({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"})
```

```python
webpage = requests.get(url,headers=Headers)
webpage
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
soup
```

```python
links = soup.find_all("a", attrs={'class':'sc-1dun5hk-0 cOiOrj'})
links
```

```python
link = links[0].get('href')
link
```

```python
property_list = 'https://www.realtor.com/' + link
property_list
```

```python
new_webpage = requests.get(property_list, headers=Headers)
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
new_soup
```

# Extracting Data

```python
# Extract Property Tittle
new_soup.find('h3', attrs={'class':'lbxu17-2 qvrsN description-title'}).text
```

```python
# Extract Property Type
new_soup.find('div', attrs={'class':'sc-12iqlu8-2 dPeBQf basicInfoValue'}).text
```

```python
# Extract List ID
new_soup.find('span', attrs={'class':'listing-id'}).text
```
